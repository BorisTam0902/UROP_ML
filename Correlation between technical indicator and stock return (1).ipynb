```python
import csv
```

```python
open_price = {}
close_price = {}
high_price = {}
low_price = {}
volume = {}
ma_15days = {}
rsi = {}
daily_return = {}
macd = {}
signal = {}
for i in range(1,10000):
    if i < 10:
        string = "000" + str(i) + ".HK.csv"
    elif i < 100:
        string = "00" + str(i) + ".HK.csv"
    elif i < 1000:
        string = "0" + str(i) + ".HK.csv"
    else:
        string = str(i) + ".HK.csv"
    try :
        open(string)
    except FileNotFoundError:
        continue
    with open(string) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        next(csv_reader)
        open_price[i] = []
        close_price[i] = []
        high_price[i] = []
        low_price[i] = []
        volume[i] = []
        ma_15days[i] = []
        rsi[i] = []
        daily_return[i] = []
        macd[i] = []
        signal[i] = []
        for row in csv_reader:
            open_price[i].append(row[3])
            close_price[i].append(row[2])
            high_price[i].append(row[4])
            low_price[i].append(row[5])
            volume[i].append(row[7])
            daily_return[i].append(row[9])
            ma_15days[i].append(row[16])
            rsi[i].append(row[17])
            macd[i].append(row[24])
            signal[i].append(row[25])
        daily_return[0] = 0
        for j in range(15):
            ma_15days[j] = 0
            rsi[j] = 0
        for j in range(25):
            macd[j] = 0
        for j in range(33):
            signal[j] = 0
        for j in range(len(open_price)):
            open_price[i][j] = float(open_price[i][j])
            close_price[i][j] = float(close_price[i][j])
            high_price[i][j] = float(high_price[i][j])
            low_price[i][j] = float(low_price[i][j])
            volume[i][j] = float(volume[i][j])
            daily_return[i][j] = float(daily_return[i][j])
            ma_15days[i][j] = float(ma_15days[i][j])
            rsi[i][j] = float(rsi[i][j])
            macd[i][j] = float(macd[i][j])
            signal[i][j] = float(signal[i][j])
```

```python
# store different length of rsi14 value
rsi_dict = {}
for i in range(1,51):
    rsi_collection = []
    for j in range(len(rsi)):
        daily_value = []
        for k in range((j-i if (j-i) > 0 else 0),j):
            daily_value.append(rsi[k])
        rsi_collection.append(daily_value)
    rsi_dict[i] = rsi_collection
print(rsi_dict[10])
```

```python
# store different length of macd value
macd_dict = {}
for i in range(1,51):
    macd_collection = []
    for j in range(len(rsi)):
        daily_value = []
        for k in range((j-i if (j-i) > 0 else 0),j):
            daily_value.append(macd[k])
        macd_collection.append(daily_value)
    macd_dict[i] = macd_collection
print(macd_dict[10])
```

```python
signal_dict = {}
for i in range(1,51):
    signal_collection = []
    for j in range(len(signal)):
        daily_value = []
        for k in range((j-i if (j-i) > 0 else 0),j):
            daily_value.append(signal[k])
        signal_collection.append(daily_value)
    signal_dict[i] = signal_collection
print(signal_dict[10])
```

```python
# Store different windows of daily return
one_day_return = {}
five_day_return = {}
ten_day_return = {}
for i in range(len(daily_return)):
    one_day_return[i] = daily_return[i]
for i in range(len(daily_return)):
    five_day_actual_return = 1
    for j in range(5):
        five_day_actual_return = five_day_actual_return * (1 + daily_return[(i+j) if (i+j) <= len(daily_return) - 1 else len(daily_return) - 1])
    five_day_return[i] = five_day_actual_return
for i in range(len(daily_return)):
    ten_day_actual_return = 1
    for j in range(10):
        ten_day_actual_return = ten_day_actual_return * (1 + daily_return[(i+j) if (i+j) <= len(daily_return) - 1 else len(daily_return) - 1])
    ten_day_return[i] = ten_day_actual_return
print(ten_day_return)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
macd_curve = []
new_daily_return = daily_return[33:900]
for i in range(33,900):
    macd_curve.append(macd[i]-signal[i])
print(len(macd_curve))
print(len(new_daily_return))
```

```python
x_axis = np.array(macd_curve).reshape(-1,1)
y_axis = np.array(new_daily_return)
```

```python
model = LinearRegression()
model.fit(x_axis, y_axis)
r_sq = model.score(x_axis, y_axis)
```

```python
print('alpha:', model.intercept_)
print('beta:', float(model.coef_))
print('r squared value:', r_sq)
```
